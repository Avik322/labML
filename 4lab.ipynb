```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Генерация случайных точек
n_samples = 500
X, _ = make_blobs(n_samples=n_samples, centers=2, random_state=42)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c='blue', marker='o', label='Точки')
plt.title('Генерация случайных точек')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

```python
from sklearn.mixture import GaussianMixture

# EM-алгоритм с использованием Gaussian Mixture Model
def em_algorithm(X, n_components=2):
    gmm = GaussianMixture(n_components=n_components, random_state=42)
    gmm.fit(X)
    return gmm
    
gmm = em_algorithm(X)

# Визуализация
x, y = np.meshgrid(np.linspace(X[:, 0].min(), X[:, 0].max(), 100),
                   np.linspace(X[:, 1].min(), X[:, 1].max(), 100))

pos = np.dstack((x, y))
pdf = np.exp(gmm.score_samples(pos.reshape(-1, 2))).reshape(x.shape)

plt.figure(figsize=(8, 6))
plt.contourf(x, y, pdf, levels=20, cmap='Blues', alpha=0.7)
plt.scatter(X[:, 0], X[:, 1], c='red', marker='o', label='точки')
plt.title('EM-алгоритм: восстановленная плотность вероятности')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import KernelDensity

# KDE
def kde_algorithm(X, bandwidth=0.5):
    kde = KernelDensity(bandwidth=bandwidth)
    kde.fit(X)
    return kde

kde = kde_algorithm(X)

# Визуализация
pdf_kde = np.exp(kde.score_samples(pos.reshape(-1, 2))).reshape(x.shape)

plt.figure(figsize=(8, 6))
plt.contourf(x, y, pdf_kde, levels=20, cmap='Blues', alpha=0.7)
plt.scatter(X[:, 0], X[:, 1], c='green', marker='o', label='точки')
plt.title('Оценка плотности ядра: восстановленная плотность вероятности')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

```python
# Метрополиса-Гастингса
def metropolis_hastings(pdf_func, initial_point, n_steps=1000, step_size=0.1):
    samples = [initial_point]
    current_point = initial_point

    for _ in range(n_steps):
        proposal_point = current_point + np.random.normal(0, step_size, size=current_point.shape) # Пробуем сгенерировать новую точку

        p_current = pdf_func(current_point) # Вероятность принятия нового состояния
        p_proposal = pdf_func(proposal_point)

        acceptance_ratio = min(1, p_proposal / p_current)
        
        if np.random.rand() < acceptance_ratio: # Принимаем предложение с вероятностью acceptance_ratio
            current_point = proposal_point

        samples.append(current_point)
    
    return np.array(samples)

# Восстановленная плотность вероятности из GMM
def gmm_pdf(point):
    return np.exp(gmm.score_samples([point]))

# Генерация точек с помощью M-H
initial_point = np.array([0, 0])
mh_samples = metropolis_hastings(gmm_pdf, initial_point)

# Визуализаци
plt.figure(figsize=(8, 6))
plt.scatter(mh_samples[:, 0], mh_samples[:, 1], c='purple', marker='o', label='сгенерированно с помощью M-H')
plt.title('Выборка Метрополис-Гастингс')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()

```

```python
# Метод Гибсона
def gibson_sampling(pdf_func, initial_point, n_steps=1000, step_size=0.1):
    samples = [initial_point]
    current_point = initial_point

    for _ in range(n_steps):
        proposal_point = current_point + np.random.normal(0, step_size, size=current_point.shape)
        
        if np.random.rand() < pdf_func(proposal_point) / pdf_func(current_point):
            current_point = proposal_point

        samples.append(current_point)
    
    return np.array(samples)

# Генерация точек с помощью Гибсона
gibson_samples = gibson_sampling(gmm_pdf, initial_point)

# Визуализация результатов
plt.figure(figsize=(8, 6))
plt.scatter(gibson_samples[:, 0], gibson_samples[:, 1], c='orange', marker='o', label='сгенерированно с помощью Гибсона')
plt.title('Выборка Гибсона')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt

# Функция для вычисления KL расстояния
def kl_divergence(p, q, grid_points):
    p_values = p(grid_points.T)  # Плотность для распределения p (передаем .T)
    q_values = q(grid_points.T)  # Плотность для распределения q (передаем .T)
    
    # Обработаем случай, когда плотности равны нулю для некоторых точек
    epsilon = 1e-10  # Малое значение для защиты от нулевых плотностей
    p_values = np.maximum(p_values, epsilon)
    q_values = np.maximum(q_values, epsilon)
    
    return np.sum(p_values * np.log(p_values / q_values))

# Пример данных
X = np.random.randn(100, 2)  # 100 точек в 2D
mh_samples = np.random.randn(100, 2)  # Генерация случайных точек методом M-H
gibson_samples = np.random.randn(100, 2)  # Генерация случайных точек методом Gibson

# Оценка плотности для исходных и сгенерированных точек
kde_original = gaussian_kde(X.T)  # KDE для исходных точек
kde_mh = gaussian_kde(mh_samples.T)  # KDE для сгенерированных точек методом M-H
kde_gibson = gaussian_kde(gibson_samples.T)  # KDE для сгенерированных точек методом Gibson

# Создание сетки точек для плотности
x, y = np.meshgrid(np.linspace(X[:, 0].min(), X[:, 0].max(), 100),
                   np.linspace(X[:, 1].min(), X[:, 1].max(), 100))

grid_points = np.vstack([x.flatten(), y.flatten()]).T  # Сетка точек для плотности

# Вычисление KL расстояния
kl_mh = kl_divergence(kde_original, kde_mh, grid_points)
kl_gibson = kl_divergence(kde_original, kde_gibson, grid_points)

print(f"KL Расхождение между исходными образцами и образцами MH: {kl_mh}")
print(f"KL Расхождение между оригинальными семплами и сэмплами Гибсона: {kl_gibson}")
```
